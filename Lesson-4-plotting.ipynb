# `plot()` method 
`pandas.Dataframe` has a built in method `plot` for plotting.
Default behavior of plot without specifying paramters: 

- creates one line plot for each of the columns with numeric data 
- sets the values of the indec as the x axis values 


```python
#a mess
df.plot()
```

We can make a plot of one columnn against another by using plot 
df.plot(x = `x_values_column)`, y = `y_values_column')

Example

```python
#change in glacial volume per eyar in europe


df.plot(x = 'year', y = 'europe')
```

Basis customization specifying other arguments plot()

Example: 

```python
df.plot(x = 'year', 
        y = 'europe',
       title = 'Change in glacial volume per year in Europe',
       xlabel = 'Year',
       ylabel = 'Change in glacial volume (km3)',
        color = 'green')
```

# Multiple line plots 

We want to graph the change in gacial volume in the Artic and Alaska Update thee arguments:
- 'y': a list of column names that will be plotted against x
- 'color': specifiy the color of each column with a dictionary '{col1:`color1`, col2:`color2`}'

Examples:


```python
colors = {'arctic': '#F48FB1',
         'alaska': 'yellow'} #specify nicer colros

df.plot(x='year',
       y = ['arctic', 'alaska'],
       color = colors)
```

Create separate plots using tehh `subset` argument  to `True`

```python
df.plot(x='year',
       y=['arctic', 'alaska'], 
       color=colors,
       subplots=True)
```

# Scatterplots 
Example: visually compare the flipper length against body mass


```python
penguins.plot(kind = 'scatter',
             x = 'flipper_length_mm',
              y = 'body_mass_g',
              alpha = 0.4 #controls transparency of points 
             )
```

## Bar plot 

set `kind='bar'` in the `plot(`) method 

example: we want to get data about teh 1- penguins witht eh lowest body mass and visualize eit 

1. find penguins with smallest body mass using `nsmallest()` and method for series

```python
smallest = penguins.body_mass_g.nsmallest(10) #n smallest, 10 smallest
smallest
```

```python
#order it from smallest to biggest, 
```

```python
# sort_values order values in descending order 
smallest = penguins.body_mass_g.nsmallest(10).sort_values() #parentheis calls fucniton and gets output 
smallest
```

```python
#plot smalelst body masses as bars 
smallest.plot(kind = 'bar')
```
